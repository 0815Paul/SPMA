```python
# Importing the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random

from datetime import datetime
```

```python
PATH_IN = '../data/input/'

filepaths = [
            PATH_IN + '/assets/chp.csv',
            PATH_IN + '/assets/chp_operation1_2.csv',
            PATH_IN + '/assets/chp_operation2_3.csv',
        ]

chp_data_dict = {}
for filepath in filepaths:
    data = pd.read_csv(filepath, index_col=0)
    chp_data_dict[filepath] = data
```

```python
chp_data = chp_data_dict['../data/input//assets/chp.csv']
chp_op12_data = chp_data_dict['../data/input//assets/chp_operation1_2.csv']     
chp_op23_data = chp_data_dict['../data/input//assets/chp_operation2_3.csv']
```

```python
# Thermal Efficiency over Heat
eta_th_2 = chp_op12_data.loc[2,'eta_th']
eta_th_1 = chp_op12_data.loc[1,'eta_th']
heat_1 = chp_op12_data.loc[1,'heat']
heat_2 = chp_op12_data.loc[2,'heat']

# Plotten
plt.figure(figsize=(10, 6))
plt.plot([heat_1, heat_2], [eta_th_1, eta_th_2], marker='o', linestyle='-', color='b')
plt.title('Thermal Efficiency over Heat')
plt.xlabel('Heat')
plt.ylabel('Thermal Efficiency (eta_th)')
plt.grid(True)
plt.show()
```

```python
# Gas Consumption over Heat
gas_2 = heat_2 / eta_th_2
gas_1 = heat_1 / eta_th_1

# Plotten
plt.figure(figsize=(10, 6))
plt.plot([heat_1, heat_2], [gas_1, gas_2], marker='o', linestyle='-', color='b')
plt.title('Gas Consumption over Heat')
plt.xlabel('Heat')
plt.ylabel('Gas Consumption')
plt.grid(True)
plt.show()
```

```python
eta_el_2 = chp_op12_data.loc[2,'eta_el']
eta_el_1 = chp_op12_data.loc[1,'eta_el']

# Electrical Efficiency over Heat
plt.figure(figsize=(10, 6))
plt.plot([heat_1, heat_2], [eta_el_1, eta_el_2], marker='o', linestyle='-', color='b')
plt.title('Electrical Efficiency over Heat')
plt.xlabel('Heat')
plt.ylabel('Electrical Efficiency (eta_el)')
plt.grid(True)
plt.show()
```

```python
# Power over Heat
power_2 = eta_el_2 * gas_2
power_1 = eta_el_1 * gas_1

# Plotten
plt.figure(figsize=(10, 6))
plt.plot([heat_1, heat_2], [power_1, power_2], marker='o', linestyle='-', color='b')
plt.title('Power over Heat')
plt.xlabel('Heat')
plt.ylabel('Power (el)')
plt.grid(True)
plt.show()
```
